```python
# Imports the os module, which provides functions for interacting with the operating system.  
# This module is used for file and directory manipulations.
import os

# This will display the directory where the Jupyter server was launched, which is where the notebook is typically saved.
notebook_path = os.getcwd()
print("Current Jupyter Notebook path:", notebook_path)

# Imports the pandas library, commonly used for data manipulation and analysis.  
# It is imported as pd which is a standard alias in the Python community.
import pandas as pd
```

```python
# Defines a function named process_excel_file that takes one parameter, file_path, which is the path to an Excel file that will be processed.
def process_excel_file(file_path):
        
    # Reads the Excel file located at file_path into a DataFrame named df. 
    # The DataFrame is a 2-dimensional data structure that can hold data in a tabular format.
    df = pd.read_excel(file_path)

    # prints the first five and last five rows of each excel file
    display(df)
    
    # Drops the first row (index 1) from the DataFrame, essentially deleting the 1st row.
    df = df.drop(1)

    # Checks if there are at least 11 columns in the DataFrame. If the condition is met, this line drops the 11th column (index 10).
    # By specifying the axis=1 (column axis), the drop() method removes the specified column.
    if len(df.columns) >= 11:
        df = df.drop(df.columns[10], axis=1)

    # This for loop iterates over each column in the DataFrame. It checks if any entry in the column contains a comma and does not contain .jpeg.
    # If the condition is satisfied, this line moves the first column with a comma to the front and breaks the loop.
    for col in df.columns:
        if df[col].astype(str).str.contains(',').any() and not df[col].astype(str).str.contains('.jpeg').any():
            df = df[[col] + [c for c in df.columns if c != col]]
            break

    # Look for "https://www.munchies.com" in any column and move the first such column to the second position
    for col in df.columns:
        if df[col].astype(str).str.contains('https://www.munchies.com').any():
            df = df[[df.columns[0], col] + [c for c in df.columns if c != df.columns[0] and c != col]]
            break

    # Look for column with "% OFF" and empty cell, move the first such column to the third position
    for col in df.columns:
        if df[col].astype(str).str.contains('% OFF').any() and df[col].isna().any():
            df = df[[df.columns[0], df.columns[1], col] + [c for c in df.columns if c != df.columns[0] and c != df.columns[1] and c != col]]
            break

    for col in df.columns:
        if df[col].astype(str).str.contains('').any() and df[col].isna().any():
            df = df[[df.columns[0], df.columns[1], col] + [c for c in df.columns if c != df.columns[0] and c != df.columns[1] and c != col]]
            break
    
    for col in df.columns:
        if df[col].astype(str).str.contains('Preorder').any() and df[col].isna().any():
            df = df[col, [df.columns[0], df.columns[1]] + [c for c in df.columns if c != df.columns[0] and c != df.columns[1] and c != col]]
            break

    for col in df.columns:
        if df[col].astype(str).str.contains('Preorder').any() and df[col].isna().any():
            df = df[[df.columns[0], df.columns[1],col] + [c for c in df.columns if c != df.columns[0] and c != df.columns[1] and c != col]]
            break

    

    # Remove rows that only contain "All Restaurants"
    df = df[~(df == "All Restaurants").all(axis=1)]

    # Rename columns
    df.columns = [f"column_{i+1}" for i in range(len(df.columns))]

    return df
```

```python
# Defines a function which takes two parameters: input_directory and output_directory, 
# These specify where to read Excel files from and where to save the processed files.
def download_processed_files(input_directory, output_directory):
    
    # This creates the output directory if it doesn't already exist. 
    # The exist_ok=True flag prevents an error if the directory already exists.
    os.makedirs(output_directory, exist_ok=True)

    # This lists all files in the input directory and filters them to include only those that end with .xlsx.
    file_list = os.listdir(input_directory)
    excel_files = [file for file in file_list if file.endswith('.xlsx')]

    # This loop iterates over each Excel file found. It constructs the full input and output file paths using os.path.join. 
    # This creates paths that are compatible with the operating system.
    for excel_file in excel_files:
        input_file_path = os.path.join(input_directory, excel_file)
        output_file_path = os.path.join(output_directory, excel_file)

        # For each Excel file, it calls the process_excel_file function and saves the processed DataFrame to the output directory using to_excel.
        # The index=False parameter prevents the DataFrame index from being saved as an additional column in the output file.
        
        df = process_excel_file(input_file_path)
        df.to_excel(output_file_path, index=False)

    # Print statements provide feedback that the processing is complete and files have been saved.
    print("Processing completed.")
    print("Files downloaded to the specified output directory.")
```

```python
# Directory where your Excel files are located
directory = "/Users/talha.yamin/DataSolutions 360/Python Project/output"

# Check if the directory exists
if os.path.exists(directory):
    # List all Excel files in the directory
    file_names = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith(".xlsx")]
    print("Excel files to combine:", file_names)  # Print the list of file names

    # Create an empty DataFrame to store the combined data
    combined_data = pd.DataFrame()

    # Initialize a counter for the number of processed files
    processed_file_count = 0

    # Loop through the Excel files and sheets
    for file_name in file_names:
        print(f"Processing file: {file_name}")
        excel_file = pd.ExcelFile(file_name)
        for sheet_name in excel_file.sheet_names:
            print(f" - Processing sheet: {sheet_name}")
            try:
                # Read the specific columns (2 and 7) from each sheet and drop completely empty rows
                data = pd.read_excel(file_name, sheet_name, usecols=[1, 6])
                
                # Remove any rows with all NaN values
                data = data.dropna(how="all")
                
                # Strip whitespace from each cell if it's a string
                data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
                
                # Append the cleaned data
                combined_data = pd.concat([combined_data, data], ignore_index=True)
                
            except Exception as e:
                print(f"   Error processing {sheet_name}: {e}")  # Print the error message
            
        # Increment the processed file counter
        processed_file_count += 1

    # Rename columns in the combined DataFrame
    combined_data.columns = ["Vendor Name", "URL Link"]
    
    # Save the combined data to a new Excel file in the specified directory
    output_file = os.path.join(directory, "combined_data.xlsx")
    combined_data.to_excel(output_file, index=False)
    print(f"Combined data saved to: {output_file}")

    # Print the number of files processed
    print(f"Total files processed: {processed_file_count}")
else:
    print(f"The directory '{directory}' does not exist.")
```

```python
#For a faster but less accurate solution, we perform the following fuzzy match.
# A module for fuzzy string matching, which allows for approximate matching of strings.
from rapidfuzz import process

# A library for creating progress bars in Python loops, useful for tracking the progress of long-running processes.
from tqdm import tqdm

# Step 1: Convert combined_data.xlsx to combined_data.csv
combined_data_excel_file = "/Users/talha.yamin/DataSolutions 360/Python Project/output/combined_data.xlsx"
combined_data_csv_file = "/Users/talha.yamin/DataSolutions 360/Python Project/output/combined_data.csv"

# Load the Excel file and save it as a CSV
combined_data = pd.read_excel(combined_data_excel_file)
combined_data.to_csv(combined_data_csv_file, index=False)

# Step 2: Load the vendors.csv file
csv_file = "/Users/talha.yamin/DataSolutions 360/Python Project/vendors.csv"
vendors_data = pd.read_csv(csv_file)
vendors_data['vendor_name_lower'] = vendors_data['vendor_name'].str.strip().str.lower()

# Step 3: Load the combined_data.csv file
combined_data = pd.read_csv(combined_data_csv_file)

# Initialize an empty list to store the results
result_data = []

# Step 4: Iterate through the rows in combined_data and match with vendors_data
for index, row in tqdm(combined_data.iterrows(), total=len(combined_data), desc="Processing Rows"):
    vendor_1 = row['Vendor Name']  # Adjust this based on your actual column name
    
    # Perform fuzzy matching for vendor_1 only
    matches_vendor_1 = process.extract(vendor_1, vendors_data['vendor_name_lower'], limit=5)

    # Use a set to track already added vendor names to avoid duplicates
    matched_names = set()

    for match in matches_vendor_1:
        if match[1] > 80:  # Threshold for similarity score
            matched_row = vendors_data[vendors_data['vendor_name_lower'] == match[0]].iloc[0]
            # Check if we have already added this vendor_name to avoid duplicates
            if matched_row['vendor_name'] not in matched_names:
                result_data.append({
                    'Vendor_1': vendor_1,
                    'vendor_name': matched_row['vendor_name'],
                    'vendor_code': matched_row['vendor_code']
                })
                matched_names.add(matched_row['vendor_name'])

# Step 5: Create a DataFrame from the result_data list
result_df = pd.DataFrame(result_data)

# Step 6: Save the result to a CSV file
result_output_file = "/Users/talha.yamin/DataSolutions 360/Python Project/output/result_output.csv"
result_df.to_csv(result_output_file, index=False)

print("Fuzzy matching completed. Results saved to:", result_output_file)
```
